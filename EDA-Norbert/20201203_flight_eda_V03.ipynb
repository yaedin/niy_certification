```python
def my_info(df):
  info = pd.DataFrame(df.count(), columns=["filled"])
  info["filled %"] = round(info["filled"] / len(df), 4) * 100
  info["nunique"] = df.nunique()
  info["dtypes"] = df.dtypes
  info["uniques"] = np.nan
  for idx, row in info.iterrows():
    info.loc[idx, "uniques"] = str(list(df[idx].unique()))
  return info
```

# 1. Which airports do we have the most delays?

```python
flight_01 = flight_01.loc[(flight_01['DIVERTED'] == 0) & (flight_01['CANCELLED'] == 0)].copy()
```

```python
# Definition of a delay with DELAY_LIMIT. All flight with a delay of 15 Minutes are a delayed flight
# Generate new column DELAY with start value 0 
# Fill column DELAY with all flights as 1 depending on the definition 

delay_limit = 30

flight_01['DELAY'] = 0

flight_01.loc[flight_01['ARR_DELAY_NEW'] > delay_limit, 'DELAY']=1
```

```python
# Number of the Flights in the definition of delay
flight_01['DELAY'].sum()
```

```python
#Generate a dataframe with only delayed flight

flight = flight_01['DELAY']==1

flight_delay = flight_01[flight]

display((flight_delay.shape))

display(flight_delay.describe())
```

```python
# View on the delays of incoming flights in minutes. If an aircraft arrives earlier, the minutes are set to 0

display(flight_delay['ARR_DELAY_NEW'].describe())



```

```python
#Determination of the airport with the most delays in minutes

flight_airport_min = flight_delay.groupby(['ORIGIN_CITY_NAME', 'ORIGIN_AIRPORT_ID'])['ARR_DELAY_NEW'].sum().sort_values(ascending=False)
flight_airport_min.head(25)
```

```python
flight_airport_min.head(10).plot(kind='bar');
plt.title("AIRPORTS WITH THE HIGHEST DELAYED FLIGHTS IN MINUTES")
plt.xlabel("AIRPORTS")
plt.ylabel("Number of delayed minutes")
plt.show()
```

----> Chicago is the airport with the most delays in minutes


```python
#Determination of the airport with the most flights

flight_airport_flights = flight_01.groupby(['ORIGIN_CITY_NAME', 'ORIGIN_AIRPORT_ID'])['FLIGHTS'].count().sort_values(ascending=False)
flight_airport_flights.sum()
```

```python
# Consideration of the number of delays for incoming flights from each carrier

flight_carrier_num = flight_delay.groupby(['OP_CARRIER'])['ARR_DELAY_NEW'].count().sort_values(ascending=False)

display(flight_carrier_num)
display((flight_carrier_num).sum())

flight_carrier_num.head(10).plot(kind='bar');
plt.title("NUMBER OF DELAYED INCOMING FLIGHT FROM EACH CARRIER")
plt.xlabel("Carrier")
plt.ylabel("Number of delayed flights")
plt.show()


```

---> OO (Skywest Airlines/Delta Connection/United Express (USA) Skywest) has the most of delayed flights


```python

#  Consideration of the delays in minutes for incoming flights from each carrier

flight_delay_carrier_min = flight_delay.groupby(['OP_CARRIER', 'DELAY'])['ARR_DELAY_NEW'].sum().sort_values(ascending=False)


display(flight_delay_carrier_min)
display((flight_delay_carrier_min).sum())
display(flight_delay_carrier_min.count())


flight_delay_carrier_min.head(10).plot(kind='bar');
plt.title("MINUTES OF DELAYED INCOMING FLIGHT FROM EACH CARRIER")
plt.xlabel("Carrier")
plt.ylabel("Number of delayed flights")
plt.show()
```

---> OO (Skywest Airlines/Delta Connection/United Express (USA) Skywest) has the most of minutes of delays

```python
flight_delay_carrier_number=flight_delay.groupby(['OP_CARRIER', 'DELAY'])['FLIGHTS'].sum()
display(flight_delay_carrier_number.count())
display(flight_delay_carrier_number.sum())
```

```python
# share of delayed flights in total flights per carrier
flight_carrier  = (flight_delay.groupby(['OP_CARRIER', 'DELAY'])['ARR_DELAY_NEW'].sum()/
                   flight_delay.groupby(['OP_CARRIER', 'DELAY'])['FLIGHTS'].sum()).sort_values(ascending=False)

display(flight_carrier.sum())
display(flight_carrier)

flight_carrier.head(15).plot(kind='bar');
plt.title("DELAYED MINUTES PER FLIGHTS FOR EACH CARRIER")
plt.xlabel("Carrier")
plt.ylabel("delayed minutes")
plt.show()
```

---> 33% of all flights from the airline C5 are delayed, however the airline only has only 3858 flights.

sorted by the number of flights:
WN 7,9%
AA 10,1%
DL 7,4% delay

```python
# check avg of delay per flight in minutes
flight_delay_carrier_min.sum()/flight_delay_carrier_number.sum()
```

---> On average, the airline OO has a delay of arriving flights of 122 minutes for all flights.

```python
# Determination of the delay in minutes for incoming flights from each carrier depending on the airport

flight_dest_min = flight_delay.groupby(['OP_CARRIER', 'DEST_CITY_NAME'])['ARR_DELAY_NEW'].sum().sort_values(ascending=False)

display(flight_dest_min)
display((flight_dest_min).sum())

flight_dest_min.head(10).plot(kind='bar');
plt.title("DELAY IN MINUTES FOR INCOMING FLIGHTS FROM EACH CARRIER DEOENDING ON THE AIRPORT")
plt.xlabel("Carrier")
plt.ylabel("number of minutes")
plt.show()
```

---> OO (Skywest Airlines/Delta Connection/United Express (USA) Skywest) has the most of minutes of delays in Chicago

```python
# Determination of the delay in number of flights for incoming flights from each carrier depending on the airport

flight_dest_num = flight_delay.groupby(['OP_CARRIER', 'DEST_CITY_NAME'])['FLIGHTS'].count().sort_values(ascending=False)

display(flight_dest_num)
display((flight_dest_num).sum())

flight_dest_num.head(10).plot(kind='bar');
plt.title("NUMBER OF DELAYS FOR INCOMING FLIGHTS FROM EACH CARRIER DEPENDING ON THE AIRPORT")
plt.xlabel("Carrier")
plt.ylabel("number of delays")
plt.show()
```

```python
(flight_dest_min/flight_dest_num).sort_values(ascending=False)
```

---> If you look at the total number of airlines' flights, AA (...) flights in Atlanta have the most delays

```python
# Determination of the delay for incoming flights from each carrier

flight_dest = (flight_delay.groupby(['OP_CARRIER', 'DEST_CITY_NAME'])['ARR_DELAY_NEW'].sum()/
               flight_delay.groupby(['OP_CARRIER', 'DEST_CITY_NAME'])['FLIGHTS'].count()).sort_values(ascending=False)

display(flight_dest.head(25))

flight_dest.head(25).plot(kind='bar');
plt.title("MINUTES for DELAYED FLIGHTS IN TOTAL PER CARRIER")
plt.xlabel("Carrier")
plt.ylabel("number of delayed minutes")
plt.show()
```
