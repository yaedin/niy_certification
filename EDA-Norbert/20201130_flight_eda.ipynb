```python
# Helperfunction


def info(dataframe):
  print('dataframe: =============================================================')  
  display(dataframe.head(5))
  print('SHAPE: =============================================================')
  display(dataframe.shape)
  print('info: =============================================================')
  display(dataframe.info())
  print('describe: =============================================================')
  display(dataframe.describe())
  print('dtypes: =============================================================')
  display(dataframe.dtypes)
  print('isnull: =============================================================')
  display(dataframe.isnull().sum())
  print('sample: =============================================================')
  display(dataframe.sample(5))
  print('=============================================================')
```
